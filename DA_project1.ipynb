```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('retail_sales_dataset.csv')

# Show first 5 rows
df.head()
```

```python
# Check missing values
df.isnull().sum()
```

```python
df.describe()
```

```python
print(df.columns)
df.columns = df.columns.str.strip()
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.strip()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
sales_trend = df.groupby('Date')['Total Amount'].sum()

plt.figure(figsize=(10,5))
sales_trend.plot()
plt.title("Sales Over Time")
plt.show()
```

```python
top_products = df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Products")
plt.show()
```

```python
product_category_sales = df.groupby('Product Category')['Total Amount'].sum()

product_category_sales.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title("Sales by Product Category")
plt.ylabel('') # Hide the default 'Total Amount' y-label for pie chart
plt.show()
```

```python
gender_sales = df.groupby('Gender')['Total Amount'].sum()

gender_sales.plot(kind='bar')
plt.title("Sales by Gender")
plt.show()
```

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.show()
```
